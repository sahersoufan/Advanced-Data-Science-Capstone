```python
df = pd.read_csv('../data/WranglingDone.csv', index_col=[0])
```

```python
group = df.groupby(['Area'])['MeterPrice'].mean().sort_values(ascending=False)
qcu = pd.qcut(group, q = 10)
qcuList = qcu.unique().tolist()
qcuAreaPriceDic = {}
for index, value in enumerate(qcuList):
    qcuAreaPriceDic[value] = 'area'+ str(index+1)
```

```python
features = {'Gifts' : 0, 'Mortgage' : 0, 'Sales' : 0, 'area1' : 0, 'area10' : 0, 'area2' : 0, 'area3' : 0,
       'area4' : 0, 'area5' : 0, 'area6' : 0, 'area7' : 0, 'area8' : 0, 'area9' : 0, '1 B/R' : 0, '2 B/R' : 0,
       '3 B/R' : 0, 'OtherRooms' : 0, 'Studio' : 0, '1' : 0, 'B' : 0, 'OtherParking' : 0, 'P' : 0,
       'Buj Khalifa Dubai Mall Metro Station' : 0, 'Business Bay Metro Station' : 0,
       'Creek Metro Station' : 0, 'Damac Properties' : 0, 'Dubai Internet City' : 0,
       'First Abu Dhabi Bank Metro Station' : 0, 'Jumeirah Lakes Towers' : 0,
       'Nakheel Metro Station' : 0, 'OtherMetro' : 0, 'Rashidiya Metro Station' : 0,
       'Sharaf Dg Metro Station' : 0, 'City Centre Mirdif' : 0, 'Dubai Mall' : 0,
       'Ibn-e-Battuta Mall' : 0, 'Mall of the Emirates' : 0, 'Burj Al Arab' : 0,
       'Burj Khalifa' : 0, 'Downtown Dubai' : 0, 'Motor City' : 0, 'OtherLandmark' : 0,
       'Sports City Swimming Academy' : 0}
```

```python
#Functions:
def getArea(ar):
    areaValue = qcu[ar]
    return qcuAreaPriceDic[areaValue]

def getParking(pr):
    parkingList = ['B', 'P', '1']
    pr = str(pr)[0]
    return 'OtherParking' if str(pr) not in parkingList else pr

def setFeatures(feat):
    # set features
    feat['Area'] = getArea(feat['Area'])
    feat['Parking'] = getParking(feat['Parking'])
    
    currentQueary = features.copy()
    for i in feat:
        if feat[i] in  currentQueary.keys():
            currentQueary[feat[i]] = 1
    return pd.Series(currentQueary)

def predict(x):
    x = setFeatures(x)
    x = ca.transform(x.to_numpy().reshape(1, -1))
    pre = xgboost.predict(x)
    print(np.expm1(pre))
    return (np.expm1(pre)).tolist()
    
# x = { 
# 'Transaction Type' : 'OtherTransactionType', 
# 'Area' : 'Al Bada', 
# 'Room' : '1 B/R', 
# 'Parking' : 's-g', 
# 'Nearest Metro' : 'Buj Khalifa Dubai Mall Metro Station', 
# 'Nearest Mall' : 'Mall of the Emirates', 
# 'Nearest Landmark' : 'Burj Khalifa',
# 'propsize' : '1'
# }
# predict(x)
```

```python
from flask_cors import CORS
app = Flask(__name__)
cors = CORS(app, resources={r"/predict/*": {"origins": "*"}})


@app.route('/')
def hello():
    return 'hello, world!'

@app.route('/predict', methods=['POST'])
def getPredict():
    try:
        data = request.get_json(force=True)
        print(data)
        d = jsonify(predict(data))
        print(d)
        return d
    except:
        print_exc()
        return jsonify(success = False)

if __name__ == '__main__':
    app.run()

```
